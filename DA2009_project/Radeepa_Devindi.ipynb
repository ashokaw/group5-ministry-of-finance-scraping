## **Extract from Statements and Remarks of the Ministry of finance website.**

```python
import requests

url = "https://www.treasury.gov.lk/web/statements-remarks/section/secretary-statements-remarks"
headers = {'User-Agent': 'Mozilla/5.0'}  
response = requests.get(url, headers=headers)
html = response.text
response
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html.parser")

print(soup.prettify())
```

##### **find all div elements**

```python
div_tags = soup.find_all("div",{'class':'header'})
len(div_tags)
```

#### **Extract the data in the Statement and Remarks in the Ministry of Finance** ####

```python
for i in range(len(div_tags)):
    var1 = div_tags[i].text
    print(var1)
```

```python

import re

content_section = soup.find("div", class_="page-cards__section")

all_text = content_section.get_text()

names = re.findall(r'(?:Mr\.|Ms\.|Mrs\.|Dr\.)\s+[A-Z][a-z]+\s+[A-Z][a-z]+', all_text)

speech_titles = []
for a_tag in content_section.find_all('a'):
    speech_titles.append(a_tag.get_text().strip())

print("\n=== NAMES ===")
for name in set(names):
    print(name)

print(f"\nTotal speeches found: {len(speech_titles)}") #To find all number of speeches

print("\n=== SPEECH TITLES ===")
for i, title in enumerate(speech_titles[:10], 1):  # Show first 10
    print(f"{i}. {title}")
```

#### **To get all links of the documents in Statements and remarks** ####

```python
links = [a['href'] for a in content_section.find_all('a') if a.has_attr('href')]

for link in set(links):
    print(link)

print(f"Total links found: {len(links)}")
```
