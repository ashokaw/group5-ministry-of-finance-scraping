**Opening a PDF and getting data in all pages**

```python
import pdfplumber

with pdfplumber.open("PDFs/NPP 2025 English .pdf") as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        text = page.extract_text()
        print(f"\n--- Page {page_num} ---\n")
        print(text)
```
